```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
tips = pd.read_csv('week10/Sample_Data_for_Activity.csv')
tips.head()
tips.info()
sns.displot(tips['Normal_Distribution'], kde=True)
```

```python
sns.displot(tips['Normal_Distribution'], kde=True, bins=100)
```

```python
sns.jointplot(x='Normal_Distribution', y='Uniform_Distribution', data=tips, kind='scatter')
```

```python
sns.jointplot(x='Normal_Distribution', y='Uniform_Distribution', data=tips, kind='reg')
```

```python
sns.jointplot(x='Normal_Distribution', y='Uniform_Distribution', data=tips, kind='hex')
```
